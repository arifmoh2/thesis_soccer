## Matches
#### Only looking at domestic leagues

How do we best store this messy data?

```python
# loading the match data
matches={}
nations = ['Italy','England','Germany','France','Spain']
for nation in nations:
    with open('../DataJson/matches/matches_%s.json' %nation) as json_data:
        matches[nation] = json.load(json_data)
```

Have a MATCH dataframe:

    wyID
    status
    roundId
    gameweek
    team_home
    team_away
    seasonId
    date
    winner
    venue
    label
    duration
    competitionId

And also a MATCH_DETAILS folder where each entry is class of the following structure:

    teamA (int)
    teamB (int)

    lineupA (list of dict(playerId, owngoal, redcard, goal))
    benchA (list of dict(playerId, owngoal, redcard, goal))

    lineupB (list of dict(playerId, owngoal, redcard, goal))
    benchB (list of dict(playerId, owngoal, redcard, goal))

    subs(list of dict(in, out, minute)) 

```python
def get_home_away(teamsData):
    team1, team2 = teamsData.keys()
    if teamsData[team1]['side'] == 'home' and teamsData[team2]['side'] == 'away':
        return int(team1), int(team2)
    elif teamsData[team2]['side'] == 'home' and teamsData[team1]['side'] == 'away':
        return int(team2), int(team1)
    else:
        print('ERROR')
        
        
def get_goals(teamsData, spec):
    team1, team2 = teamsData.keys()
    if teamsData[team1]['side'] == 'home' and teamsData[team2]['side'] == 'away':
        home = team1
        away = team2
    elif teamsData[team2]['side'] == 'home' and teamsData[team1]['side'] == 'away':
        home = team2
        away = team1
    else:
        print('ERROR')
        
    if spec == 'regular':
        return teamsData[home]['score'], teamsData[away]['score']
    elif spec == 'HT':
        return teamsData[home]['scoreHT'], teamsData[away]['scoreHT']
    elif spec == 'ET':
        return teamsData[home]['scoreET'], teamsData[away]['scoreET']
    elif spec == 'P':
        return teamsData[home]['scoreP'], teamsData[away]['scoreP']
    else:
        print('ERROR')
        return None
    
    
matches_df = list()   
for nation in nations:
    for i, match in enumerate(matches[nation]):
        #print(nation, i)
        
        match_entry = dict()
        match_entry['wyId'] = match['wyId']
        match_entry['nation'] = nation
        match_entry['status'] = match['status']
        match_entry['roundId'] = match['roundId']
        match_entry['gameweek'] = match['gameweek']
        match_entry['seasonId'] = match['seasonId']
        match_entry['date'] = match['date']
        match_entry['winner'] = match['winner']
        match_entry['venue'] = match['venue']
        match_entry['label'] = match['label']
        match_entry['duration'] = match['duration']
        match_entry['competitionId'] = match['competitionId']
        
        match_entry['team_home'], match_entry['team_away'] = get_home_away(match['teamsData'])
        match_entry['goals_home'], match_entry['goals_away'] = get_goals(match['teamsData'], 'regular')
        match_entry['goalsHT_home'], match_entry['goalsHT_away'] = get_goals(match['teamsData'], 'HT')
        match_entry['goalsET_home'], match_entry['goalsET_away'] = get_goals(match['teamsData'], 'ET')
        match_entry['goalsPen_home'], match_entry['goalsPen_away'] = get_goals(match['teamsData'], 'P')

        
        matches_df.append(pd.Series(match_entry))

matches_df = pd.concat(matches_df, axis=1).T.set_index(['wyId', 'nation']).sort_index(axis=1).sort_index()
matches_df.to_pickle(os.path.join(ROOT, 'processed_data', 'matches_df.pkl.gz'))
matches_df.to_csv(os.path.join(ROOT, 'processed_data', 'matches_df.csv'))
matches_df
```

```python
class MatchDetails:
    def __init__(self, teamsData):
        team1, team2 = teamsData.keys()
        if teamsData[team1]['side'] == 'home' and teamsData[team2]['side'] == 'away':
            home = team1
            away = team2
        elif teamsData[team2]['side'] == 'home' and teamsData[team1]['side'] == 'away':
            home = team2
            away = team1
        else:
            print('ERROR')
            
            
        self.home = int(home)
        self.away = int(away)
        
        
        self.goals_home, self.goals_away = get_goals(teamsData, 'regular')
        self.goalsHT_home, self.goalsHT_away = get_goals(teamsData, 'HT')
        self.goalsET_home, self.goalsET_away = get_goals(teamsData, 'ET')
        self.goalsPen_home, self.goalsPen_away = get_goals(teamsData, 'P')
        
        self.lineup_home =  teamsData[home]['formation']['lineup'] if teamsData[home]['hasFormation'] else None
        self.lineup_away =  teamsData[away]['formation']['lineup'] if teamsData[away]['hasFormation'] else None
        
        self.bench_home =  teamsData[home]['formation']['bench'] if teamsData[home]['hasFormation'] else None
        self.bench_away =  teamsData[away]['formation']['bench'] if teamsData[away]['hasFormation'] else None
        
        self.substitutions_home = teamsData[home]['formation']['substitutions'] if teamsData[home]['hasFormation'] else None
        self.substitutions_away = teamsData[away]['formation']['substitutions'] if teamsData[away]['hasFormation'] else None
```

```python
for nation in nations:
    for i, match in enumerate(matches[nation]):
        details = MatchDetails(match['teamsData'])
        #details.to_pickle(os.path.join(ROOT, 'processed_data', 'match_details', '{}.pkl.gz'.format(match['wyId'])))
        
        pickle.dump(details, open(os.path.join(ROOT, 'processed_data', 'match_details', '{}.pkl.gz'.format(match['wyId'])),
                                  "wb" ) )
```
